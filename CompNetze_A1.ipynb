**Plotten Sie das Spektrum eines amplitudenmodulierten Signals mit Datenrate D=1 kBit/s und Frequenz des Trägersignals ftr = 10 kHz**

Lösung:

Frequenz des quadratischen Signals Fs0 bestimmen: 

Fs0=1/Tp; Tp=2*Tb Tb=1/D D=1000b/s  

-> Tb=1/1000=1ms -> Tp=2ms -> Fs0=1/2000=500Hz

Import libraries

```python
import numpy as np
import matplotlib.pyplot as plot
from scipy import signal
from scipy.fft import fft, rfft, rfftfreq, fftfreq
```

Trägersignal visualisieren

```python
# Zeit von 0s bis 0,004s -> x-Werte fürs Diagram
t=np.linspace(0,0.004,1000) 

# Frequenz von Trägersignal
Ftr = 10000 

# y-Werte fürs Diagram
Str = np.cos(t*Ftr) 

# Title fürs Diagram
plot.title('Trägersignal')

# Name für x-Achse
plot.xlabel('Zeit')

# Name für y-Achse
plot.ylabel('Amplitude ')

# Visualisieren das Netz
plot.grid(True, which='both')

plot.axhline(y=0, color='black')

plot.plot(t, Str)
```

Signal visualisieren

```python
# Frequenz des Signals
Fs0 = 500 

# y-Werte fürs Diagram
S=0.5*signal.square(2 * np.pi * t * Fs0) + 0.5

# Title fürs Diagram
plot.title('Signal')

# Name für x-Achse
plot.xlabel('Zeit')

# Name für y-Achse
plot.ylabel('Amplitude ')

# Visualisieren das Netz
plot.grid(True, which='both')

plot.axhline(y=0, color='black')

plot.plot(t, S)
```

Amplituden-modulierten Signal visualisieren

```python
AmS=S*Str

# Title fürs Diagram
plot.title('Amplitudenmoduliertes Signal')

# Name für x-Achse
plot.xlabel('Zeit')

# Name für y-Achse
plot.ylabel('Amplitude ')

# Visualisieren das Netz
plot.grid(True, which='both')

plot.axhline(y=0, color='black')

plot.plot(t, AmS)
```
